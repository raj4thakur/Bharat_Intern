```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x_train=df.drop(columns=["quality"])
y_train=df["quality"]
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_train,x_test,y_train,y_test=train_test_split(x_train,y_train,random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score,accuracy_score,mean_squared_error,confusion_matrix
```

```python
model_reg=LogisticRegression()
model_reg.fit(x_train,y_train)
```

```python
model_R=RandomForestClassifier()
model_R.fit(x_train,y_train)
```

```python
pd.Series(model_R.feature_importances_,index=df.drop("quality",axis=1).columns).plot(kind="bar")
```

```python
pred=model_R.predict(x_train)
accuracy_score(y_train,pred)
```
